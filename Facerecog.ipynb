```python
import cv2
faceCascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
image = cv2.imread("sampic.jpeg")
```

```python
from matplotlib import pyplot as plt
plt.imshow(image)
```

```python
RGB_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

```python
GR_img = cv2.cvtColor(RGB_img,cv2.COLOR_BGR2GRAY)
plt.imshow(GR_img)
```

Detect coordinates of faces


```python
faces=faceCascade.detectMultiScale( GR_img, scaleFactor=1.1,minNeighbors=5, minSize=(30,30))
print(faces)
```
